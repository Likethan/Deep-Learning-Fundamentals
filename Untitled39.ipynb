```python
# Activation Functions: ReLU and Sigmoid
import numpy as np
import matplotlib.pyplot as plt

# Input values
x = np.linspace(-10, 10, 1000)

# ReLU Function
def relu(x):
    return np.maximum(0, x)

# Sigmoid Function
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

# Calculate activations
y_relu = relu(x)
y_sigmoid = sigmoid(x)

# Plotting
plt.figure(figsize=(12, 5))

# ReLU Plot
plt.subplot(1, 2, 1)
plt.plot(x, y_relu, color='blue')
plt.title('ReLU Activation Function')
plt.xlabel('x')
plt.ylabel('ReLU(x)')
plt.grid(True)

# Sigmoid Plot
plt.subplot(1, 2, 2)
plt.plot(x, y_sigmoid, color='green')
plt.title('Sigmoid Activation Function')
plt.xlabel('x')
plt.ylabel('Sigmoid(x)')
plt.grid(True)

plt.tight_layout()
plt.show()
```
